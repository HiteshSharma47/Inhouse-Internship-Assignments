# Q.1) Create a CSV file for address book, CSV file should have column for name, address, mobile, email. Insert 2-3 dummy data entered by user.


```python
import csv 

#Create and Insert Data In CSV File\

l = [
    ["Name","Address","Mobile","Email"],
    ["Hitesh","Jaipur","8675438985","123@gmail.com"],
    ["Ajay","Jodhpur","9890657843","someone@gmail.com"],
]

with open("address_book.csv","w") as f:
    writer = csv.writer(f)

    for rows in l:
        writer.writerow(rows)

#Read Data From CSV

with open("address_book.csv","r") as f:
    writer = csv.reader(f)

    for rows in writer:
        print(rows)
```

# Q.3) Create 2-3 tables


```python
import sqlite3

conn = sqlite3.connect("Practice.db")

#Table 1
sql = """
create table student(
id integer primary key autoincrement,
name varchar(50),
email varchar(50),
)
"""

#Table 2
sql1 = """
create table faculty(
id integer primary key autoincrement,
name varchar(50),
email varchar(50)
)
"""

conn.execute(sql)
conn.execute(sql1)
conn.close()


```

# Q.4) Insert some records


```python
import sqlite3

conn = sqlite3.connect("Practice.db")
#For Table 1
ins = """
insert into student(name,email)values("Hitesh","123@gmail.com")
"""
ins1 = """
insert into student(name,email) values("Ajay","someone@gmail.com")
"""
conn.execute(ins)
conn.execute(ins1)

conn.commit()
#For Table 2
ins3 = """
insert into faculty(name,email)values("Mr Rajesh","1234@gmail.com")
"""
ins4 = """
insert into faculty(name,email) values("Mr Vikram","someoneelse@gmail.com")
""" 
conn.execute(ins3)
conn.execute(ins4)

conn.commit()
conn.close()
```

# Q.5) Perform different select operations


```python
import sqlite3

conn = sqlite3.connect("Practice.db")

#From Table 1
print("For Student Table:\n")
data = conn.execute("select * from student")

for m in data:
    print(m)

print("\nFor Faculty Table:\n")

data1 = conn.execute("select * from faculty")

for n in data1:
    print(n)

conn.close()
```

# Q.6) Update some data


```python
import sqlite3

conn = sqlite3.connect("Practice.db")

#for Table 1
conn.execute("update student set name = 'Vansh' where name ='Hitesh' ")
conn.commit()

data = conn.execute("select * from student")

print("Student Table After Update:\n")
for m in data:
    print(m)
#for Table 2
conn.execute("update faculty set email = 'noone@gmail.com' where email ='1234@gmail.com' ")
conn.commit()

data = conn.execute("select * from faculty")

print("\nFaculty Table After Update:\n")
for n in data:
    print(n)
    
conn.close()
```

# Q.7) Delete some data

```python
import sqlite3

conn = sqlite3.connect("Practice.db")

#for Table 1
conn.execute("delete from student where id = 1")
conn.commit()

data = conn.execute("select * from student")

print("Student Table After delete:\n")
for m in data:
    print(m)
#for Table 2
conn.execute("delete from faculty where name = 'Mr Vikram' ")
conn.commit()

data = conn.execute("select * from faculty")

print("\nFaculty Table After Delete:\n")
for n in data:
    print(n)
    
conn.close()
```
