# Q.1) Practise Pandas Series
# Create a Pandas Series from Dictionary 
# Create a Pandas Series from Lists 
# Access the elements of a Series in Pandas

```python
# Create Panda Series from Dictionary

import pandas as pd

data = {
    'Messi':10,
    'Ronaldo':7,
    'Neymar':11,
    'Lewandowski':9
}

df = pd.Series(data)
print("-----Series of Dictionary-----\n")
print(df)

#From List

data2 = ['Messi','Ronaldo','Neymar','Lewandowski']

print("\n-----Series of Lists-----\n")
df2 = pd.Series(data2)

print(df2)

#Access the Elements

print("\n-----Access using index-----\n")
print("From Dictionary Series :",df['Messi'])
print("From List Series :",df2[0])

print("\n-----Access using iloc-----\n")
print("From Dictionary Series :\n",df.iloc[0:3])
print("\nFrom List Series :\n",df2.iloc[0:2])
```

# Q.2) DataFrames
# Make a Pandas DataFrame with a two-dimensional Python list 
# Create DataFrame from Python dict 
# Create Pandas dataframe using list of lists 
# Create a Pandas dataframe using list of tuples 
# Create a Pandas DataFrame from List of Dicts 


```python
#DataFrame Using 2D Python List

import pandas as pd

data = [
    [100,100,100],
    [95,100,90],
    [99,89,94],
    [85,90,98]
]
print("-----Pandas DataFrame Using 2D Python List-----\n")

df = pd.DataFrame(data,index=['Messi:','Ronaldo:','Neymar','Lewandowski'])
df.columns = ['Java','Python','C++']
print(df)

#DataFrame Using Python Dict
data = {
    'Players':['Messi','Ronaldo','Neymar','Lewandowski'],
    'Jersey Number':[10,7,11,9]
}

df = pd.DataFrame(data,index=[1,2,3,4])

print("\n-----Pandas DataFrame Using Python Dict-----\n")
print(df)

#Using List of Lists

data = [
    ['Messi',10],
    ['Ronaldo',7],
    ['Neymar',11],
    ['Mbappe',9]
]

df = pd.DataFrame(data,columns = ['Players','Jersey'],index = [1,2,3,4])

print("\n-----Pandas DataFrame Using List Of Lists-----\n")
print(df)

#Using List Of Tuples

data = [
    ('C',2),
    ('C++',4),
    ('Python',1),
    ('Java',3)
]

df = pd.DataFrame(data,columns = ['Language','Ranking'],index = [1,2,3,4])

print("\n-----Pandas DataFrame Using List Of Tuples-----\n")
print(df)

#Using List Of Dicts

data=[
    {'Language':'C++','Ranking':4},
    {'Language':'C','Ranking':2},
    {'Language':'Python','Ranking':1},
    {'Language':'Java','Ranking':3},
]

df = pd.DataFrame(data,index = [1,2,3,4])

print("\n-----Pandas DataFrame Using List Of Dictionaries-----\n")
print(df)

    
    
```

# Q.3)Data iteration
# Different ways to iterate over rows in Pandas Dataframe 
# Selecting rows in pandas DataFrame based on conditions 
# Select any row from a Dataframe using iloc[]
# Limited rows selection with given column
# Drop rows from the dataframe based on certain condition applied on a column 
# Insert row at given position in Pandas Dataframe 
# Create a list from rows in Pandas dataframe 


```python
#Diff Ways Of Iteration Over Rows in DataFrame

import pandas as pd

data = {
    "Player": ["Messi", "Ronaldo", "Neymar"],
    "Goals": [10, 7, 11]
}

df = pd.DataFrame(data)

#Simple Iteration
print("\n1).-----Simple Iteration-----\n")
for i in range(len(df)):
    print(i,df['Player'][i],df['Goals'][i])

#Using iterrows
print("\n-----Using iterrows-----\n")
for index , row in df.iterrows():
    print(index , row['Player'],row['Goals'])

#Selecting rows in pandas DataFrame based on conditions

print("\n-----Selection Based on Conditions-----\n")
result = df[df['Goals'] > 8]
print(result)

#Select any row from a Dataframe using iloc[]

print("\n-----Selection Based on iloc()-----\n")
print(df.iloc[0])
print()
print(df.iloc[[0]])

#Limited rows selection with given column using loc()

print("\n-----Limit Row Selection using a given Column using loc()-----\n")
print(df.loc[0:2,['Goals']])

#Drop rows from the dataframe based on certain condition applied on a column

print("\n-----Drop Rows based on conditions on column-----\n")
result = df.drop(df[df['Goals'] < 9].index)
print(result)

#Insert row at given position in Pandas Dataframe
result = df.values.tolist()

result.insert(1,['Mbappe',9])

df1 = pd.DataFrame(result,columns = df.columns)
print("\n-----Insert rows at given position in DataFrame-----\n")
print(df1)

#Create a list from rows in Pandas dataframe

print("\n-----Create a List from rows in DataFrame-----\n")
rows = df.values.tolist()
print(rows)
```
