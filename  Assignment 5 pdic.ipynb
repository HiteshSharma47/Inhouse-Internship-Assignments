# 1) Write a Python program to repeat a tuple three times using the * operator.

```python
t = (1, 2, 3)

print("\n---Original Tuple---\n")
print(t)

print("\n---Tuple Repeated 3 Times---\n")

res = t * 3
print(res)
```

#  2) Write a Python program to join three separate tuples into one new tuple using the + operator.

```python
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

print("\n---Original Tuples---\n")

print(t1)
print(t2)
print(t3)

print("\n---Joined Tuple---\n")

res = t1 + t2 + t3
print(res)
```

# 3) Write a Python program to check whether a specific element exists inside a tuple using the in keyword.

```python
t = (1, 2, 3, 4)

print("\n---Tuple---\n")
print(t)

element = int(input("Enter The Element To Check:"))

print("\n---Checking Element---\n")

if element in t:
    print(element, "exists in tuple")
else:
    print(element, "does not exist in tuple")
```

# 4) Write a Python program to calculate the total, highest value, and lowest value from a tuple of integers without using the built-in sum(), max(), and min() functions.

```python
t = (12, 45, 8, 23, 67, 3)

print("\n---Tuple---\n")
print(t)

sum = 0
max = t[0]
min = t[0]

for i in t:
    sum += i

    if i > max:
        max = i

    if i < min:
        min = i

print("\n---Results---\n")

print("Sum =", sum)
print("Max =", max)
print("Min =", min)
```

# 5) Write a Python program to filter a tuple . n = (3, 14, 7, 22, 9, 41, 18, 5), keep only values greater than 10 

```python
t = (3, 14, 7, 22, 9, 41, 18, 5)

print("\n---Original Tuple---\n")
print(t)

result = ()

for i in t:
    if i > 10:
        result += (i,)

print("\n---Filtered Tuple (>10)---\n")
print(result)
```

# 6) Write a Python program to determine how many elements are in a set without using the built-in len() function. s = {"cat", "dog", "bird", "fish"}

```python
s = {"cat", "dog", "bird", "fish"}

print("\n---Set---\n")
print(s)

count = 0

for item in s:
    count += 1

print("\n---Number Of Elements---\n")
print(count)
```

# 7) Write a Python program to combine two sets into one, containing all unique elements from both sets.

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

print("\n---Set 1---\n")
print(s1)

print("\n---Set 2---\n")
print(s2)

print("\n---Combined Set---\n")

res = s1.union(s2)

print(res)
```

# 8) Write a Python program to find all elements that are common to both sets. s1 = {1, 2, 3, 4} s2 = {3, 4, 5, 6}

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print("\n---Set 1---\n")
print(s1)

print("\n---Set 2---\n")
print(s2)

print("\n---Common Elements---\n")

res = s1.intersection(s2)

print(res)
```

# 9) Write a Python program to find all elements that are in either set but not both

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print("\n---Set 1---\n")
print(s1)

print("\n---Set 2---\n")
print(s2)

print("\n---Elements Present In Either Set But Not Both---\n")

res = s1.union(s2) - s1.intersection(s2)

print(res)
```
