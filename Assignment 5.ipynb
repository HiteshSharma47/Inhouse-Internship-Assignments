# Q.1) Study the open weather API show more data in your API calling program

```python
import requests

#call OpenWeather API

def weather(city):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=e43abfcda2944a6979fc7bed52d818ea&units=metric"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        print(f"Temp of {city} :\n",data['main']['temp'])
        print(f"Maximum temp of {city} :\n",data['main']['temp_max'])
        print(f"Humidity of {city} :\n",data['main']['humidity'])
        print(f"Country of the {city} :\n",data['sys']['country'])
        
    except requests.exceptions.RequestException as e:
        print(e)


#User defined City
a = input("Enter City Name :")
weather(a)
```

# Q.2) Try building today's game yourself


```python
#ROCK,PAPER,SCISSORS Game

import random

choices = ["rock", "paper", "scissors"]

while True:
    print("1. Start Game")
    print("2. Exit")

    choice = input("Enter your choice: ")

    if choice == "2":
        print("Game exited")
        break

    elif choice == "1":
        user_score = 0
        computer_score = 0

        while user_score < 3 and computer_score < 3:
            user = input("Enter rock, paper or scissors: ").lower()
            computer = random.choice(choices)

            print("Computer chose:", computer)

            if user == computer:
                print("It's a tie")

            elif (
                (user == "rock" and computer == "scissors") or
                (user == "paper" and computer == "rock") or
                (user == "scissors" and computer == "paper")
            ):
                user_score += 1
                print("You win this round")

            elif user in choices:
                computer_score += 1
                print("Computer wins this round")

            else:
                print("Invalid input")
                continue

            print("Your Score:", user_score)
            print("Computer Score:", computer_score)
            print()

        if user_score == 3:
            print("You won the game")
        else:
            print("Computer won the game")

    else:
        print("Invalid choice")
```
