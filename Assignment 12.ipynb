# Q.1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]

```python
print("\n-----Replace NaN with 0 and Interchange 3 rows and 3 columns-----")
import numpy as np

arr = np.array([
    [6, -8, 73, -110],
    [np.nan, -8, 0, 94]
])

print("\n---Array---\n", arr)

res = np.nan_to_num(arr, nan = 0)
print("\n---After Replacing---\n", res)

print("\n---Swapping Rows---\n")
res[[0,1],:] = res[[1,0],:]
print(res)

print("\n---Swapping 1st and 3rd Column---\n")
res[:,[0,2]] = res[:,[2,0]]
print(res)
```

# Q.3) Replace NaN values with average of columns 

```python
print("\n-----Replace NaN values with Avg of Columns-----")

arr = np.array([
    [1, 2, np.nan],
    [4, np.nan, 6],
    [7, 8, 9]
])

print("\n---Array---\n", arr)
col_mean = np.nanmean(arr, axis=0)

nan_pos = np.where(np.isnan(arr))

arr[nan_pos] = col_mean[nan_pos[1]]

print("\n---After Replacing---\n",arr)
```

# Q.4) Replace negative value with zero in numpy array using replace

```python
print("\n-----Replace Negative Values With Zero-----")

arr = np.array([6, -8, 73, -110, 5])

print("\n---Array---\n",arr)

#NumPy Arrays Dont Have Replace, We Use where
arr = np.where(arr < 0, 0, arr)

print("\n---After Replacing---\n",arr)
```

# Q.5) Study the following program import numpy as np 
# Create a numpy 1d-arrays arr1 = np.array([3, 4]) arr2 = np.array([1, 0]) 
# Find average of NumPy arrays avg = (arr1 + arr2) / 2 , print("Average of NumPy arrays:\n", avg) 
# Calculate average mean median mode values of two NumPy 2d-arrays 

```python
import numpy as np

print("\n---Average of 2 1D NumPy Arrays---\n")
arr1 = np.array([3,4])
arr2 = np.array([1,0])

print("\n---Array 1---\n", arr1)
print("\n---Array 2---\n", arr2)

#Average

avg = (arr1 + arr2) / 2

print("\nAverage Of arr1 and arr2 : ", avg)

#Mean,Median and Mode
print("\n---Mean, Median, Mode of 2 2D NumPy Arrays---")

arr1 = np.array([
    [1,1],
    [3,4]
])
arr2 = np.array([
    [5,6],
    [2,8]
])

print("\n---Array 1---\n", arr1)
print("\n---Array 2---\n", arr2)

print("\nMean of Combined arr1 and arr2 : \n", np.mean([arr1,arr2]))
print("\nMedian of Combined arr1 and arr2 : \n", np.median([arr1,arr2]))

print("\n---Flatten Array---")
flat_arr = np.concatenate([arr1.reshape(-1),arr2.reshape(-1)])
print(flat_arr)

l = np.zeros(flat_arr.max() + 1, dtype = int)
print("\n---Frequency Array---\n", l)

for x in flat_arr:
    l[x] += 1 

print("\n---After Frequency Count---")
print(l)

for x in range(l.size - 1):
    if l[x] == l.max():
        mode = x
print("\nMode of the Combined arr1 and arr2 :\n", mode)
```

# Q.6) Solve the following equation using linalg() and inverse Matrix method x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17 

```python
# using linalg()
print("\n-----Solving Using linalg()-----")
arr1 = np.array([
    [1, -2, 3],
    [-9, 3, -1],
    [2, -5, 5]
])

arr2 = np.array([17,17,17])

print("\n---Literals---\n", arr1)
print("\n---Constants---\n", arr2)

res = np.linalg.solve(arr1,arr2)

print("\n---Solution---\n", res.round())

#using inverse Matrix method

print("\n-----Solving Using Inverse Matrix Method-----")

arr1 = np.array([
    [1, -2, 3],
    [-9, 3, -1],
    [2, -5, 5]
])

arr2 = np.array([17,17,17])

inv_arr1 = np.linalg.inv(arr1)

print("\n---Inversed Array---\n", inv_arr1)

res = np.dot(inv_arr1, arr2)

print("\n---Solution---\n", res.round())
```
