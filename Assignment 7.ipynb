# Q.1) Convert a series of date-strings to a timeseries


```python
import pandas as pd

#list of Dates

l = ['11-03-2006','12-02-2004','7-01-1967','7-09-2007']

df = pd.Series(l)
print("\n---LIST---\n")
print(df)

#to timeseries

new_df = pd.to_datetime(df)
print("\n---TimeSeries---\n")
print(new_df)
```

# Q.2)Create two DataFrames, df1 and df2, with a common column (e.g., 'ID'). 

# Perform an inner merge on this common column and display the resulting DataFrame.
# Perform a left join of df1 and df2 on the 'ID' column. Explain how missing values are handled in the resulting DataFrame. Right Join and Index-Based Join.
# Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys.

```python
#Creating 2 DataFrames

import pandas as pd

one = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['A1','A2','A3','A4'],
    'Marks':[95,99,89,90]
})
two = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['A1','A2','B3','B4'],
    'Marks':[95,78,87,99]
})

print("\n---Initial DataFrame---\n")

print(one)
print()
print(two)

print("\n---Inner Merge---\n")

res = one.merge(two, on = 'ID', how = 'inner')
print(res)

print("\n---Left Join---\n")

res = one.merge(two, on = 'ID', how = 'left')
print(res)
print("\n--Missing Values Handeled By NaN--\n")

print("\n---Right Join---\n")

res = one.merge(two, on = 'ID', how = 'right')
print(res)

print("\n---- INDEX BASED JOIN ----")

one_index = one.set_index('ID')
two_index = two.set_index('ID')

index_join = one_index.join(two_index, how='inner',lsuffix = 'left', rsuffix = 'right')

print(index_join)

#Using Multiple Keys

print("\n---Multiple Keys---\n")
res = one.merge(two, on = ['ID','Name'], how = 'inner')
print(res)
    
```
