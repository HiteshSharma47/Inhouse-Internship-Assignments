# Q.1)Write a Python function to find the maximum of three numbers.

```python
def maxi(a, b, c):
    if a >=b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
a= int(input("Enter a :"))
b= int(input("Enter b :"))
c= int(input("Enter c :"))

print(f"\nMax Element From {a}, {b} and {c} is :", maxi(a, b, c))
                
```

# Q.2) Write a Python function that takes a list and returns a new list with distinct elements from the first list.


```python
def distinct(l):
    x = []
    for a in l:
        if a not in x:
            x.append(a)
    return x

a = int(input("Enter List Range:"))
l = []
for x in range(a):
    b = int(input("Enter List Elements:"))
    l.append(b)
print("\nDistinct List :",distinct(l))
```

# Q.3) Write a Python function to multiply all the numbers in a list.

```python
def multiply(l):
    mul = 1;
    for x in l:
        mul *= x
    return mul

a = int(input("Enter List Range:"))
l = []
for x in range(a):
    b = int(input("Enter List Elements:"))
    l.append(b)

print("\nAfter Multiplication of Each Elements :",multiply(l))
```

# Q.4) Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.


```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
x = int(input("Enter Number For Factorial:"))
while x < 0:
    print("\nInvalid Input!")
    x = int(input("Enter Again:"))
print(f"\nFactorial Of {x} is :", factorial(x)) 
    
    
```

# Q.5) Write a Python program to reverse a string.


```python
def rev(s):
    return s[::-1]

s = input("Enter String :")
print("\nReverse is :",rev(s))
```

# Q.6) Write a Python function to check whether a number falls within a given range.


```python
def f(s, e, c):
    if c <= e and c >= s:
        print(f"\n{c} is in the range of {s} and {e}")
    else:
        print(f"\n{c} is not in the range of {s} and {e}")

s = int(input("Enter The Range From :"))
e = int(input("Enter The Range To :"))
c = int(input("Enter The Number To Check:"))

f(s, e, c)
```

# Q.7) Write a Python function to Print Even Numbers from a Given List


```python
def even(l):
    b = []
    for x in l:
        if x % 2 == 0:
            b.append(x)
    return b

a = int(input("Enter List Range:"))
l = []
for x in range(a):
    c = int(input("Enter List Elements:"))
    l.append(c)
print("\n Even Number List is :",even(l))



            
            
```

# Q.8) Write a Python function that takes a number as a parameter and checks whether the number is prime or not.


```python
def prime(n):
    if n < 2:
        return False
    for x in range (2,n):
        if(n % x == 0):
            return False
    return True

n = int(input("Enter The Number :"))

if(prime(n)):
    print(f"\n{n} is a Prime Number!")
else:
    print(f"\n{n} is not Prime Number!")
```

# Q.9) Write a Python function that accepts a string and counts the number of upper and lower case letters. 


```python
def count(s):
    upper = 0
    lower = 0
    for x in s:
        if x >= 'a' and x <= 'z':
            lower += 1
        elif x >= 'A' and x <= 'Z':
            upper += 1
        else:
            pass
    print("\nNumber Of Upper Letters :", upper)
    print("\nNumber Of Lower Letters :", lower)

s = input("Enter The String : ")

count(s)
```
