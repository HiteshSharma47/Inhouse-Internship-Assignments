# Q.1) Write a python program that takes in a student name, class. It should also take in five subject marks of the students and find the total mark and percentage. Display a result in such a way that their name, class,  and percentage are printed.

```python
name = input("Enter Student name:")
section = input("Enter class name:")

maths = int(input("Enter Maths Marks:"))
dbms = int(input("Enter DBMS Marks:"))
dsa = int(input("Enter DSA Marks:"))
java = int(input("Enter Java Marks:"))
python = int(input("Enter Python Marks:"))


marks = maths + dbms + dsa + java + python
percentage = (marks / 500) * 100
print(f"\nTotal Marks of {name} is : {marks}")
print(f"Percentage of {name} is : {percentage:.2f}%\n")
print("Result:\n")
print(f"Name : {name}")
print(f"class : {section}")
print(f"Percentage : {percentage:.2f}%")
```

# Q.2) Input 2 strings concatinate them and store in another variable. then perform generally used string methods on it likelower(), upper(), title(), swapcase(), capitalize(), casefold(), center(), count(), endswith(), find(), isalnum(), isdigit(), isnumeric(), isspace(), replace()

```python
str1 = input("Enter String 1:")
str2 = input("Enter String 2:")
concatinate = str1 +" " + str2
print(f"\nConcatination of {str1} and {str2}:",concatinate)

print("\n using lower():",concatinate.lower())

print("\n using upper():",concatinate.upper())

print("\n using title():",concatinate.title())

print("\n using swapcase():",concatinate.swapcase())

print("\n using capitalize():",concatinate.capitalize())

print("\n using casefold():",concatinate.casefold())

print("\n using center():",concatinate.center(50))

print("\n using count():",concatinate.count("p"))

print("\n using endswith():",concatinate.endswith("programming"))

print("\n using find():",concatinate.find("is"))

print("\n using isalnum():",concatinate.isalnum())

print("\n using isdigit():",concatinate.isdigit())

print("\n using isnumeric():",concatinate.isnumeric())

print("\n using isspace():",concatinate.isspace())

print("\n using replace():",concatinate.replace("python","C++"))
```

# Q.3) Do practise of assignment operators

```python
print("///ARITHMETIC OPERATORS///\n")
x = 5
print(f"for x = 5: x assigns {x}\n")
x += 5
print(f"for x += 5: x assigns {x}\n")
x = 5
x -= 5
print(f"for x -= 5: x assigns {x}\n")
x = 5
x *= 5
print(f"for x *= 5: x assigns {x}\n")
x = 5
x /= 5
print(f"for x /= 5: x assigns {x}\n")
x = 5
x %= 5
print(f"for x %= 5: x assigns {x}\n")
x = 5
x //= 5
print(f"for x //= 5: x assigns {x}\n")
x = 5
x **= 5
print(f"for x **= 5: x assigns {x}")

print("\n///COMPARISON OPERATORS///\n")

x = 5
y = 10

print("for x == y:\n",x == y)
print("for x != y:\n",x != y)
print("for x > y:\n",x > y)
print("for x < y:\n",x < y)
print("for x >= y:\n",x >= y)
print("for x <= y:\n",x <= y)

print("\n///LOGICAL OPERATORS///\n")

x = 5
y = 10

print("for x < 10 and y < 20:\n", x < 10 and y < 20)
print("for x < 10 or y > 20:\n", x < 10 or y > 20)
print("for not(x < 10 and y < 20):\n", not(x < 10 and y < 20))

print("\n///IDENTITY OPERATORS///\n")

x = 10
y = 10

print("for x is y:\n", x is y)
print("for x is not y:\n", x is not y)
```

# Q.4) In your last program where you find the total and percentage of a student's marks of 5 subject, find the grade of the student using conditional statement. Eg. grade 'A' if percentage is greator than or equals to 60, 'B' for  percentage is greator than or equals to 50 and less than 60,  'C' for  percentage is greator than or equals to 40 and less than 50,  'D' for  percentage is greator than or equals to 33 and less than 40, otherwise 'Fail'

```python
name = input("Enter Student name:")
section = input("Enter class name:")

maths = int(input("Enter Maths Marks:"))
dbms = int(input("Enter DBMS Marks:"))
dsa = int(input("Enter DSA Marks:"))
java = int(input("Enter Java Marks:"))
python = int(input("Enter Python Marks:"))


marks = maths + dbms + dsa + java + python
percentage = (marks / 500) * 100
print(f"\nTotal Marks of {name} is : {marks}")
print(f"Percentage of {name} is : {percentage:.2f}%\n")
print("Result:\n")
print(f"Name : {name}")
print(f"class : {section}")
print(f"Percentage : {percentage:.2f}%")

if percentage >= 60:
    print(f"Grade of {name} : A")
elif (percentage < 60 and percentage >= 50):
    print(f"Grade of {name} : B")
elif (percentage < 50 and percentage >= 40):
    print(f"Grade of {name} : C")
else:
    print(f"Grade of {name} : Fail!")
```
